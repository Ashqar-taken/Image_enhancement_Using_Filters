### Smoothing Filters

```python
# Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img,cmap='viridis')
plt.axis('off')
```

```python
# Perform Average Filtering:

kernel = np.ones((5,5),np.float32)/25
avg_fil_img = cv2.filter2D(img,-1,kernel)

plt.imshow(avg_fil_img,cmap='viridis')
plt.axis('off')
```

```python
# Weighted Average Filtering:

kernel = np.array([[1,2,1],[2,4,2],[1,2,1]],np.float32)/16
w_avg_fil_img = cv2.filter2D(img,-1,kernel)
plt.imshow(w_avg_fil_img,cmap='viridis')
plt.axis('off')
```

```python
# Gaussian Filtering:

gaussian_fil_img = cv2.GaussianBlur(img,(5,5),0)
plt.imshow(gaussian_fil_img,cmap='viridis')
plt.axis('off')
```

```python
# Median Filtering:

med_fil_img = cv2.medianBlur(img,5)
plt.imshow(med_fil_img,cmap='viridis')
plt.axis('off')
```

```python
# Sharpening Filtering:
# Laplacian kernel

kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
sharpening_fil_img = cv2.filter2D(img,-1,kernel)
plt.imshow(sharpening_fil_img,cmap='gray')
plt.axis('off')
```

```python
# Laplacian Operator

laplacian_fil_img = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplacian_fil_img,cmap='gray')
plt.axis('off')
```
